### **Project Summary: Flipkart Customer Service Satisfaction Analysis**

In today’s digital age, online shopping has become a part of everyday life. With platforms like Flipkart handling thousands of customer queries each day, delivering a smooth and satisfying customer service experience is more important than ever. When a customer reaches out with a complaint or issue—whether it’s about a delayed delivery, a wrong item, or a basic question—the quality of the support they receive can make or break their impression of the brand.

This project focuses on exactly that: understanding what factors influence customer satisfaction during support interactions and building a model that can predict satisfaction levels from service data.

**Objective of the Project**  
The main goal was to explore what makes a customer happy or unhappy after dealing with Flipkart’s support team. More importantly, I aimed to create a machine learning model that could predict whether a customer is likely to be satisfied, based on the details of their interaction. This way, Flipkart can identify unhappy customers in advance and fix their issues before they turn into negative experiences.

**About the Dataset**  
To get started, I worked with a dataset that includes detailed records of past customer support interactions at Flipkart. It had data points like:
- The product categories involved
- Response times from agents
- The price of the product in question
- The work shift during which the complaint was handled
- And finally, the CSAT score (Customer Satisfaction Score) given by the customer, ranging from 1 (very unsatisfied) to 5 (very satisfied)

**How I Approached the Project**  
1. **Exploring the Data (EDA)**:  
I began by studying the dataset carefully—looking for common complaint trends, typical agent response times, and how different factors like time of day affected customer satisfaction. I also cleaned the data by fixing or removing incomplete entries.

2. **Data Preprocessing for ML**:  
Since machine learning models work with numbers, I converted text fields into numeric codes. I also added new features, like how long it took to close a complaint. To make sure big numbers like prices didn’t dominate the results, I used scaling to bring all features onto the same level.

3. **Model Building**:  
I tested three different models:
- `Logistic Regression` for a basic prediction baseline
- `Decision Tree` to understand decision flow based on interaction details
- `Random Forest`, an advanced ensemble model that combines many decision trees for better accuracy

4. **Model Evaluation**:  
To understand which model performed best, I didn’t just look at accuracy. I also checked:
- **Precision**: How accurate the model is when predicting dissatisfied customers
- **Recall**: How many of the truly dissatisfied customers were detected
- **F1-Score**: A balance between precision and recall

**Final Outcome: Why Random Forest Worked Best**  
Among all the models I tried, Random Forest gave the most accurate and consistent results—over 97% accuracy across all types of customer satisfaction ratings. One major plus of this model is that it tells us which features were most important in making predictions. For example, it showed that handling time and shift timings had a strong impact on how satisfied a customer felt.

**Why This Project Matters**  
If Flipkart uses a model like this, they can:
- Quickly detect unhappy customers and fix issues before they turn into bad reviews
- Train support agents better, using real data about what affects satisfaction
- Improve service decisions by focusing on what actually matters to customers

## 3. ***Data Wrangling***

### Data Wrangling Code

##### Convert Object to Datatime format

```python
# Convert 'order_date_time', 'Issue_reported at', 'issue_responded', 'Survey_response_Date' and 'connected_handling_time' columns to datetime format
datetime_cols = ['order_date_time', 'Issue_reported at', 'issue_responded']
for col in datetime_cols:
    df[col] = pd.to_datetime(df[col], format='%d-%m-%Y %H.%M', errors='coerce')

df['Survey_response_Date'] = pd.to_datetime(df['Survey_response_Date'], format='%d-%b-%y', errors='coerce')
```

```python
# Check for missing values in each column of the DataFrame
# This helps identify which columns have null or missing data that may need to be handled
print(df.isnull().sum())
```

##### Dropped Unnessesary Columns

```python
# Drop Unnessesary Columns
df.drop(['Unique id', 'Customer Remarks', 'Order_id', 'Agent_name', 'Supervisor', 'Manager'], axis=1, inplace=True)
```

##### Fill missing data

```python
# Fill missing values in 'order_date_time' column with the mode (most frequent date)
mode_order_date_time = df['order_date_time'].mode()[0]
df['order_date_time'] = df['order_date_time'].fillna(mode_order_date_time)
```

```python
# Fill missing values in 'Customer_City' column with 'Unknown_city'
df['Customer_City'].fillna('Unknown_City', inplace=True)
```

```python
# Fill missing values in 'Product_category' column with 'Unknown_Category'
df['Product_category'].fillna("Unknown_Category", inplace=True)
```

```python
# Fill missing values in 'Item_price' column with the median value
df['Item_price'].fillna(df['Item_price'].median(), inplace=True)
```

```python
# Fill missing values in 'connected_handling_time' column with the median value
df['connected_handling_time'].fillna(df['connected_handling_time'].median(), inplace=True)
```

```python
# Check for missing values in each column after data wrangling steps
df.isnull().sum()
```

##### Calculate response time in minutes

```python
# Calculate response time in minutes
df['Response_Time_Seconds'] = (df['issue_responded'] - df['Issue_reported at']).dt.total_seconds()
```

##### Calculate survey delay (days between order and survey)

```python
# Calculate survey delay (days between order and survey)
df['Survay_delay_Days'] = (df['Survey_response_Date'] - df['order_date_time']).dt.days
```

##### Categorical Encoding

```python
# Convert categorical features to category type
category_cols = df.select_dtypes(include='object').columns
df[category_cols] = df[category_cols].astype('category')
```

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

**Goal**: Define three hypothetical statements related to your dataset, formulate null and alternate hypotheses for each, perform an appropriate statistical test, and explain your choice of test.

##### **What is Hypothesis Testing?**

**Ans**: Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It involves setting up two competing statements: a `null hypothesis (H0)` and an `alternate hypothesis (H1)`.

- **Null Hypothesis (H0)**: This is the statement of no effect or no difference. It's what you try to disprove.
- **Alternate Hypothesis (H1)**: This is the statement you want to prove; it suggests there is an effect or a difference.

We will use statistical tests to calculate a `p-value`.
- **If p-value < significance level** (commonly 0.05): You reject the `null hypothesis (H0)`, meaning there's enough evidence to support the `alternate hypothesis (H1)`.
- **If p-value ≥ significance level** (commonly 0.05): You fail to reject the `null hypothesis (H0)`, meaning there's not enough evidence to support the `alternate hypothesis (H1)`.

#### **Hypothesis 1: (Comparing Numerical Data Across Categorical Groups (e.g., CSAT Score by Channel))**

#### **1. State Your research hypothesis as a `null hypothesis` and `alternate hypothesis`.**

**Hypothetical Statement**: Is there a significant difference in average `CSAT Scores` across different `channel_names`?

- **Null Hypothesis (H0)**: The mean `CSAT Score` is the same across all customer support `channel_names` (Email = Inbound = Outcall).

- **Alternate Hypothesis (H1)**: At least one `channel_name` has a significantly different mean `CSAT Score` than the others.

#### **2. Perform an appropriate statistical test.**

```python
# import required library
from scipy.stats import f_oneway

# Extract CSAT Score data for each channel group
email_csat = df[df['channel_name'] == 'Email']['CSAT Score']
inbound_csat = df[df['channel_name'] == 'Inbound']['CSAT Score']
outcall_csat = df[df['channel_name'] == 'Outcall']['CSAT Score']

# Perform one-way ANOVA test to compare mean CSAT Scores across the three channel groups
f_stat, p_value = f_oneway(email_csat, inbound_csat, outcall_csat)

# Display the ANOVA test results
print(f"F-Statistic: {f_stat:.2f}")
print(f"P-Value: {p_value:.3f}")

# Interpret the result
alpha = 0.05

if p_value < alpha:
    print("There is a significant difference in CSAT Scores across channels.")
else:
    print("There is no significant difference in CSAT Scores across channels.")
```

##### **Which statistical test have you done to obtain P-Value?**

**Ans**: I used the `ANOVA (Analysis of Variance)` statistical test to obtain the `P-value`.

##### **Why did you choose the specific statistical test?**

**Ans**: `ANOVA` was chosen because it's perfect for comparing the average of a numerical variable (`CSAT Score`) across three or more different independent channels (`channel_names`).

#### **1. State Your research hypothesis as a `null hypothesis` and `alternate hypothesis`.**

**Hypothetical Statement**: Is there a significant relationship between `connected_handling_time` and `CSAT Score`?

- **Null Hypothesis (H0)**: There is no significant relationship between `connected_handling_time` and `CSAT Score`.

- **Alternate Hypothesis (H1)**: There is a significant relationship between `connected_handling_time` and `CSAT Score`.

#### **2. Perform an appropriate statistical test.**

```python
from scipy.stats import pearsonr

temp_df = df[['connected_handling_time', 'CSAT Score']]

corr_coef, p_value = pearsonr(temp_df['connected_handling_time'], temp_df['CSAT Score'])

print(f"Pearson Correlation Coefficient: {corr_coef:.2f}")
print(f"P-Value: {p_value:.3f}")

# Interpret the result
alpha = 0.05

if p_value < alpha:
    print("There is a significant relationship between connected handling time and CSAT Score.")
else:
    print("There is no significant relationship between connected handling time and CSAT Score.")
```

#### **Hypothesis 3: Relationship Between Two Categorical Variables (e.g., Issue category and Product_category)**

#### **1. State Your research hypothesis as a `null hypothesis` and `alternate hypothesis`.**

**Hypothetical Statement**: Is there a significant relationship between the `category` of an issue and the `Product_category` it relates to?"

- **Null Hypothesis (H0)**: The `category` of an issue is independent of the `Product_category`.

- **Alternate Hypothesis (H1)**: `category` of an issue is dependent on the `Product_category`.

#### **2. Perform an appropriate statistical test.**

```python
# import required libraries
from scipy.stats import chi2_contingency
import pandas as pd

# Create a contingency table of observed frequencies between 'category' and 'Product_category'
contingency_table = pd.crosstab(df['category'], df['Product_category'])

# Perform the Chi-square test of independence on the contingency table
chi_square, p_value, dof, expected = chi2_contingency(contingency_table)

# Display the Chi-square and the P-value
print(f"Chi-square statistic: {chi_square:.2f}")
print(f"P-Value: {p_value}")

# Interpret the result
alpha = 0.05

if p_value < alpha:
    print("There is a significant relationship between category and product category.")
else:
    print("There is no significant relationship between category and product category.")
```

- **IQR (Interquartile Range) method to detect Outliers**

```python
# Detect outliers in 'connected_handling_time' using the IQR (Interquartile Range) method
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['connected_handling_time'].quantile(0.25)
Q3 = df['connected_handling_time'].quantile(0.75)

IQR = Q3 - Q1

# Compute IQR and determine lower and upper bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Since IQR is 0, all values not exactly equal to Q1/Q3 are considered outliers by this method
outliers_handling_time = df[(df['connected_handling_time'] < lower_bound) | (df['connected_handling_time'] > upper_bound)]

# Visualize the outliers in connected_handling_time using a scatter plot
plt.figure(figsize=(12,6))
plt.scatter(df.index, df['connected_handling_time'], alpha=0.3, label='All Data')
plt.scatter(outliers_handling_time.index, outliers_handling_time['connected_handling_time'], color='red', label='Outliers (IQR)')
plt.title('Outliers in Connected Handling Time (IQR Method)', fontsize=14)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Connected Handling Time', fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()

print(f"Outliers in Connected Handling Time (IQR Method): \n{outliers_handling_time.shape[0]} rows")
```
